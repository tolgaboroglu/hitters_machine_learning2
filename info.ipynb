## Çoklu Doğrusal Regresyon

```python
#Çoklu Doğrusal Regresyon  
import pandas as pd 
ad = pd.read_csv("Advertising.csv")
ad = pd.read_csv("Advertising.csv", usecols=[1,2,3,4]) # Baştaki index i değişken aldığı sütunu getirmeden import etme yöntemi.
df = ad.copy() 

X = df.drop("sales", axis=1)  #Bağımlı değişken çıkarılır. 
y = df["sales"]               #Bağımlı değişken   
```

```python
#Veriyi test ve train olarak bölme işlemi

from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.20, random_state = 42)   

training = df.copy()
```

```python
#Statsmodels ile Model Oluşturma

import statsmodels.api as sm

lm = sm.OLS(y_train, X_train)
model = lm.fit()
model.summary()

#Scikit learn ile Model Oluşturma

from sklearn.linear_model import LinearRegression

lm2 = LinearRegression()
model2 = lm2.fit(X_train, y_train)


model2.intercept_   #Sabit katsayı           B0 
model2.coef_        #Değişkenlerin katsayısı B1,B2,B3
```
